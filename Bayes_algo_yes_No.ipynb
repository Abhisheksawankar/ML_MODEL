```python
import pandas as pd

!pip install pgmpy
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
```

```python
data = pd.DataFrame({
    "CreditScore": ["High", "High", "Low", "Medium", "Low", "Medium", "High", "Low"],
    "PaymentDelay": ["No", "Yes", "Yes", "No", "Yes", "No", "No", "Yes"],
    "LoanDefault": ["No", "No", "Yes", "No", "Yes", "No", "No", "Yes"]
})
```

```python
model = DiscreteBayesianNetwork([
    ("CreditScore", "LoanDefault"),
    ("PaymentDelay", "LoanDefault")
])
```

```python
model.fit(data)
```

```python
inference = VariableElimination(model)
```

```python
result = inference.query(
    variables=["LoanDefault"],
    evidence={
        "CreditScore": "Low",
        "PaymentDelay": "Yes"
    }
)
```
